#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

#Loading Iris data

```python
iris= datasets.load_iris()
x= iris.data
y= iris.target
```

# Visualizing first 5 rows

```python
x[0:5,:]
```

```python
y[0:5,]
```

#Spliting Data

```python
from sklearn.model_selection import train_test_split
XX_train, XX_test, yy_train, yy_test,= train_test_split(x,y, test_size=0.25,random_state=100)
```

```python
XX_train.shape
```

#Importing Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
```

#Training the model

```python
# Create a Naive Bayes classifier
#classifier = MultinomialNB()    #
classifier = GaussianNB()        #   gaussian distribution
#classifier = BernoulliNB ()     #



# Fit the classifier to the training data
classifier.fit(XX_train, yy_train)
```

#Prediction

```python
yy_pred= classifier.predict(XX_test)
```

#Evaluation

```python
# Lets test the performance of the model using confusion matrix
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(yy_test,yy_pred)
print("Confusion Matrix: \n",cm)
#Performance measure accuracy
from sklearn.metrics import accuracy_score
print("Accuracy: ",accuracy_score(yy_test,yy_pred))
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt='.3f',linewidths=0.5,square=True,cmap='Pastel1');
plt.ylabel('Actual Label');
plt.xlabel('Predicted Label')
all_sample_title='Accuracy Score Naive Bias :{0}'.format(accuracy_score(yy_test,yy_pred))
plt.title(all_sample_title,size=15);
```

#Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Loading Digits Data

```python
from sklearn.datasets import load_digits
digits= load_digits()
```

#Splitting Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test,= train_test_split(digits.data,digits.target, test_size=0.25,random_state=100)
```

# Importing Digits Data from sklearn

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
```

```python
#  Digit classifier
# Create a Naive Bayes classifier
classifier = MultinomialNB()
#classifier = GaussianNB()
#classifier = BernoulliNB ()
```

#Training

```python
classifier=classifier.fit(X_train, y_train)
```

#Prediction

```python
y_pred= classifier.predict(X_test)
```

#Evaluation

```python
# Lets test the performance of the model using confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

#Confusion Matrix
cm= confusion_matrix(y_test,y_pred)
print("Confusion Matrix")
print(cm)
print("Accuracy: ",accuracy_score(y_test,y_pred))

score=classifier.score(X_test,y_test)
```

#Tumor Detection using K-Means Clustering

#Importing Libraries

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

#Loading Image

```python
imag_path = "/content/drive/MyDrive/Ai Training MetaPi/Week_5/W5_D4_clustering/tumor_image.png"
image = Image.open(imag_path).convert("L") # Convert to grayscale (working with one channel only)
```

#Converting to Gray Scale

```python
plt.imshow(image,cmap="gray")
```

*Converting to NumPy Array*

```python
# Convert PIL image to nmpy array
image_np = np.array(image)
# Reshape the image to a 1D array for K-means clustering
image_1d = image_np.reshape(-1, 1)
```

```python
image_np.shape, image_1d.shape
```

#K-means clustering

```python
n=5
kmeans = KMeans(n_clusters=n)
cluster_assignments = kmeans.fit_predict(image_1d)
```

```python
cluster_assignments.shape
```

```python
# Reshape the cluster assignments back to the image shape
clustered_image = cluster_assignments.reshape(image_np.shape)
```

```python
clustered_image.shape
```

```python
#plotting
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.imshow(image_np, cmap='gray')
plt.title('Original Image')
plt.axis('on')  # Image dimensions
plt.subplot(122)
plt.imshow(clustered_image, cmap='rainbow')
plt.title('Clustered Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

#MRI Tumor Detection using K-Means Clustering

In this project, we applied K-means clustering to detect a tumor in an MRI image. Here are the key steps and insights from the analysis:

>>-  Data Loading and Pre-processing:

We loaded an MRI image of a brain with a tumor.
The image was converted to grayscale, as we were working with a single channel (intensity values).
>>- K-Means Clustering:

We performed K-means clustering on the grayscale image to separate different regions based on intensity levels.
We chose the number of clusters as 5 (you can adjust this parameter based on the image characteristics and desired level of separation).
>>- Result Visualization:

We displayed both the original grayscale MRI image and the clustered image side by side for comparison.
The original image shows the anatomical details of the brain, while the clustered image visualizes different clusters in rainbow colors.
>>- Conclusion and Insights:

K-means clustering successfully revealed distinct regions in the MRI image, which corresponded to various structures within the brain, including the tumor. The clustered image effectively segmented the tumor region from the surrounding brain tissues. This approach provides an automated and quantitative method for identifying abnormal regions, such as tumors, in medical images. Further refinements and evaluations can be performed, and this method can be incorporated into medical image analysis pipelines to assist radiologists and medical professionals in identifying and diagnosing tumors and other anomalies.
